```python
from google.colab import drive
drive.mount('/content/drive')

path = "drive/MyDrive/"

filename = "alice_in_wonderlandch01.mp3"

from pydub import AudioSegment
from pydub.silence import split_on_silence

# split audio into several chunks
audio_data = AudioSegment.from_file(path+filename)
chunks = split_on_silence(audio_data,1000,audio_data.dBFS-20,100)
print(len(chunks))

```

```python
print(len(chunks))

import speech_recognition as sr

r = sr.Recognizer()

chunk_paths = [f"chunk_{i}.wav" for i in range(22)]
text_paths = [f"text_{i}.txt" for i in range(22)]

text_data = []
# exporting all chunks as seperate wav files
for chunk_path in chunk_paths:
    chunk = AudioSegment.from_file(path + chunk_path)
    chunk.export(path+chunk_path, format="wav")

# transcribing the text from each wav file
for chunk_path in chunk_paths:
  with sr.AudioFile(path+chunk_path) as source:
      audio = r.record(source)
      text = r.recognize_google(audio)
      text_data.append(text)

print(text_data)
print(chunk_paths)
```

```python
import nltk
nltk.download("punkt")
nltk.download("book")
from nltk import word_tokenize, sent_tokenize
import numpy as np
```

```python
# tokenize all the words from all text files in to a bag of words
w_tokens = np.array(word_tokenize(text_data[0] + text_data[1] + text_data[2]
                                 + text_data[3] + text_data[4] + text_data[5]
                                 + text_data[6] + text_data[7] + text_data[8]
                                 + text_data[9] + text_data[10] + text_data[11]
                                 + text_data[12] + text_data[13] + text_data[14]
                                 + text_data[15] + text_data[16] + text_data[17]
                                 + text_data[18] + text_data[19] + text_data[20]
                                 + text_data[21]))

print(w_tokens)
w_tokens = np.char.lower(w_tokens)
```

```python
# removing anything that isn't a word from the list
import re

is_word = []
for token in w_tokens:
  if re.search(r"[a-zA-Z]", token):
    is_word.append(token)

w_tokens = is_word
print(w_tokens)
```

```python
# removing common stopwords
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = stopwords.words("english")

# w_tokens_lower = np.char.lower(w_tokens)

not_stopwords = []
# deals with upper/lower case words
for token in w_tokens:
  if not token in stop_words and re.search(r"[^a-z]", token) or not token in stop_words and re.search(r"[^A-Z]", token):
    not_stopwords.append(token)

w_tokens = not_stopwords

print(w_tokens)
```

```python
# removing postfixes and stop words not removed by previous task i.e. 'I'
postfixes = ["n't", "'s", "'re", "'m", "na", "gon", "I",
             "You", "The", "It", "What", "She", "No", "Scene",
             "All", "That", "And", "So", "They", "'ll", "ca", "gon"]
not_postfix = []
for token in w_tokens:
  if not token in postfixes:
    not_postfix.append(token)
print(not_postfix)
w_tokens = not_postfix
```

```python
# removing words that occur less than 5 times
greater_than5 = []
for token in w_tokens:
  if w_tokens.count(token) >= 5:
     greater_than5.append(token)

w_tokens = greater_than5
print(w_tokens)
```

```python
from ast import Str
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from collections import Counter
from operator import itemgetter

# Get most common words
w_tokens_lower = np.char.lower(w_tokens)
unique, counts = np.unique(w_tokens_lower, return_counts=True)
word_counts = list(zip(unique, counts))

# sorting based on frequencies
order = np.flip(np.argsort(counts))
unique = unique[order]
counts = counts[order]

sorted_word_counts = list(zip(unique, counts))
print(sorted_word_counts)

# Credit: "inspired" by
# Huffman Coding in python
# https://www.programiz.com/dsa/huffman-coding
class TreeNode(object):
  def __init__(self, left=None, right=None, l_freq=0,r_freq=0):
    self.left = left
    self.right = right
    self.l_freq = l_freq
    self.r_freq = r_freq
    self.freq = l_freq+r_freq

  def children(self):
    return (self.left, self.right)

  def __str__(self):
    return '<Node: %s_%s (%d)>' % (self.left, self.right, (self.freq))


print("----\nInitial:", unique)

# creating the tree nodes
while len(counts) > 1:
  # extract the 2 nodes with the least amount of frequency
  u1 = unique[-1]
  u2 = unique[-2]
  c1 = counts[-1]
  c2 = counts[-2]

  # create a new node with the last one as the left node
  node = TreeNode(u1, u2, c1, c2)
  print("Creating a new node:", node)

  # print("removing",u1,"(",c1,") and ", u2,"(",c2,")")
  unique = unique[:-2]
  counts = counts[:-2]

  #put the new node back into the list
  unique = np.append(unique, node)
  counts = np.append(counts, int(c1)+int(c2))

  print("Length of unique:", len(unique))
  print("Length of counts:", len(counts))

  order = np.flip(np.argsort(counts)) #flipping to making descending
  unique = unique[order]
  counts = counts[order]

  print("----\nResorted:",unique)


print(unique[0])
```

```python
from collections import Counter
from operator import itemgetter
# generate the huffman codes
# Main function implementing huffman coding
def assign_digits(node,frequency, binString=""):
    # print("---",node,type(node),"---")
    if type(node) is TreeNode:
      # print("node",node)
      (l, r) = node.children()
      d = dict()
      d.update(assign_digits(l, node.l_freq, binString + "1")) #concatenate 0 for left
      d.update(assign_digits(r, node.r_freq, binString + "0")) #concatenate 1 for right
      return d
    else:
      print("Assigning",binString,"to",node,"(",frequency,")")
      return {node:{"bits":binString,"frequency":frequency}}

print(unique[0])
huffman_code = assign_digits(unique[0],counts[0])

huffman_data = []
for key, value in huffman_code.items():
  word = key
  bits = value["bits"]
  freq = value["frequency"]
  huffman_data.append((word, bits, freq))

print("words             Bits         Freq")
print("-----------------------------------")


sorted_huffman_data = sorted(huffman_data, key=itemgetter(2), reverse=True)

for code in sorted_huffman_data:
  print(f"{code[0]:<18} {code[1]:<12} {code[2]}")
```
